Data Cleaning

```python
# Data Cleaning
df = df.drop_duplicates()

# Check for and handle missing values
missing_values = df.isnull().sum()
```

Correlation Heatmap of Audio Features

```python
corr_matrix = df.corr(numeric_only=True)
fig = px.imshow(corr_matrix,
                labels=dict(color='Correlation'),
                x=corr_matrix.columns,
                y=corr_matrix.columns)
fig.update_layout(title='Correlation Matrix of Audio Features')
fig.show()
```

Popularity vs. Danceability

```python
fig = px.scatter(df, x='danceability_%', y='streams',
                 color='in_spotify_charts',
                 labels={'danceability_%': 'Danceability', 'streams': 'Streams'},
                 title='Popularity vs. Danceability')
fig.show()
```

Boxplot: Comparison of Spotify Chart Presence and Streams

```python
fig = px.box(df, x='in_spotify_charts', y='streams',
             labels={'in_spotify_charts': 'Spotify Chart Presence', 'streams': 'Streams'},
             title='Comparison of Spotify Chart Presence and Streams')
fig.show()
```

Bar Chart: Top Artists by Total Streams

```python
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')  # Convert 'streams' to numeric
top_artists = df.groupby('artist(s)_name')['streams'].sum().reset_index()
top_artists = top_artists.nlargest(10, 'streams')
fig = px.bar(top_artists, x='streams', y='artist(s)_name',
             labels={'streams': 'Total Streams', 'artist(s)_name': 'Artist'},
             title='Top Artists by Total Streams')
fig.show()
```

Line Chart: Temporal Trends in Danceability Over the Years

```python
temporal_trends = df.groupby('released_year')['danceability_%'].mean().reset_index()
fig = px.line(temporal_trends, x='released_year', y='danceability_%',
              labels={'released_year': 'Year', 'danceability_%': 'Danceability'},
              title='Temporal Trends in Danceability Over the Years')
fig.show()
```

Boxplot: Cross-Platform Presence vs. Total Streams

```python
platforms = ['in_spotify_charts', 'in_apple_charts', 'in_deezer_charts', 'in_shazam_charts']
df['total_platforms'] = df[platforms].sum(axis=1)
fig = px.box(df, x='total_platforms', y='streams',
             labels={'total_platforms': 'Total Platforms', 'streams': 'Streams'},
             title='Cross-Platform Presence vs. Total Streams')
fig.show()
```

Artist vs. Platform Presence and Streams

```python
fig = px.scatter(df, x='artist(s)_name', y='streams', color='total_platforms',
                 labels={'streams': 'Total Streams', 'artist(s)_name': 'Artist'},
                 title='Song Streams by Artist and Platform Presence')
fig.show()

```

Popularity vs. Energy

```python
fig = px.scatter(df, x='energy_%', y='streams', color='in_spotify_charts',
                 labels={'energy_%': 'Energy', 'streams': 'Streams'},
                 title='Popularity vs. Energy')
fig.show()
```
